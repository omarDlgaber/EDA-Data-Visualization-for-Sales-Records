```python
def safe_comma_format(x):
    try:
        # If the number is integer (such as the year), return it as it is without decimal formatting.
        if float(x).is_integer() and int(float(x)) < 3000:  # Assume the year is less than 3000
            return str(int(x))
        return format(float(x), ",.2f")
    except (ValueError, TypeError):
        return x  # If it's not a number, return it as it is
```

## 1. Top Selling Products üéØ and Top Profitable Products üí∞

### üéØ Analysis Objectives:

* What are the best-selling products?
* What are the highest-profit products?
* Are there products that excel in certain country?
* Are there products with low sales but high profits?
* Are the highest-profitable products sold through a specific channel?

```python
top_selling = sales.groupby('Item Type')['Units Sold'].sum().sort_values(ascending=False).reset_index()
top_selling.applymap(safe_comma_format)
```

```python
top_product_profit = sales.groupby('Item Type')['Total Profit'].sum().sort_values(ascending=False).reset_index()
top_product_profit.applymap(safe_comma_format)
```

## Data Visualizationüìä

```python
# Create sub-plot
fig = make_subplots(rows=1, cols=2, subplot_titles=('Top Selling Products üéØ', 'Top Profitable Products üí∞'))

# Add the first line chart (units sold)
fig.add_trace(
    go.Line(x=top_selling["Item Type"], y=top_selling["Units Sold"], name="Units Sold", marker_color='green'),
    row=1, col=1
)

# Add the second line chart (Total Profit)
fig.add_trace(
    go.Line(x=top_product_profit["Item Type"], y=top_product_profit["Total Profit"], name="Total Profit", marker_color='orange'), # ÿ™ÿ∫ŸäŸäÿ± ÿßŸÑŸÑŸàŸÜ ŸÑŸÑÿ™ŸÖŸäŸäÿ≤
    row=1, col=2
)


fig.update_layout(
    height=500,      # Adjust the height.
    width=1200,      # Adjust the witdth
    showlegend=True  # To show Units Sold and Total Profit labels
)

# Rotate the x-axis labels for both subcharts.
fig.update_xaxes(tickangle=-90, row=1, col=1)
fig.update_xaxes(tickangle=-90, row=1, col=2)

fig.show(renderer='iframe')
```

### Insights üìå
* The top-selling products were Personal Care and Clothes.

* However, the products with the highest profitability were Household and Cosmetics.

* Selling large quantities doesn't always translate to high profits; the profit margin makes a significant difference.


### Recommendation üí°
Focus on high-profit-margin products and re-evaluate the pricing and marketing strategies for products with good sales but low profitability.

| -------------------------------- | ------------------------------------------------------------- |

| `Units Sold` & `Total Profit`    | Positive correlation, but not always strong (a lot of selling does not always mean profit)|
| `Unit Cost` & `Unit Price`       | Strong positive correlation (selling price increases with cost)          |

### üîç Analysis of products that disappeared

```python
products_2016 = set(sales[sales['order_year'] == 2016]['Item Type'].unique())
products_2017 = set(sales[sales['order_year'] == 2017]['Item Type'].unique())
missing = products_2016 - products_2017
print("Products that disappeared in 2017: ", missing)
```

üîçüëÜüèª

* This means that all products that were sold in 2016 were available in 2017.
* This means that the reason for the drop is not that a product disappeared completely.

```python
plt.figure(figsize=(20, 5))

plt.subplot(1, 3, 1)
sales.groupby('order_year')['Units Sold'].sum().plot(kind='bar', title='Total Units Sold by Year')

plt.subplot(1, 3, 2)
sales.groupby('order_year')['Unit Price'].mean().plot(kind='line', marker='o', title='Average Unit Price by Year')

plt.subplot(1, 3, 3)
sales.groupby('order_year')['shipment_days_delay'].mean().plot(kind='line', marker='o', title='Avg Shipping Delay by Year')
```

üîç Analysis of the Visualizations:

üìä 1. Total Units Sold by Year
* There is a very clear and significant drop in 2017 ‚Äî from around 6.7 million units to less than 4 million.
* üëâ This confirms that the drop is real and stems from a decrease in sales volume, not just profit.

üìà 2. Average Unit Price by Year
* The unit price in 2017 remained almost the same as in 2016 and did not drop.
* üëâ This indicates that pricing is not the issue.

üöö 3. Average Shipping Delay by Year
* Interestingly, shipping delays significantly decreased in 2017!
* üëâ This is the opposite of what we'd expect ‚Äî faster shipping should usually support higher sales, not a decline.

üß† Insight Summary:

Although shipping became faster in 2017, the number of units sold dropped sharply, while prices remained nearly constant. This suggests that the decline in sales is not due to pricing or delivery issues, but may instead be linked to other factors such as market changes, reduced demand, or a cut in marketing budgets.
